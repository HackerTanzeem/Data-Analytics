```python
import matplotlib.pyplot as plt
```

```python
#line graph
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line plot')
plt.show()
```

```python
#using pandas 
import pandas as pd
df = pd.DataFrame({
    'x': [10, 8, 10, 7, 7, 10, 9, 9],
    'y': [6, 4, 2, 5, 7, 10, 9, 9]
})
print(df.head())
print(df.plot())
```

```python
#scatter graph
x=[1,2,3,6,5]
y=[2,4,9,8,10]
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot')
plt.show
```

```python
#bar graph
categories=['Cat A','Cat B','Cat C', 'Cat D']
values=[25,30,15,20]
plt.bar(categories,values)
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart')
plt.show()
```

```python
#horizontal bar
categories=['Cat A','Cat B','Cat C', 'Cat D']
values=[25,30,15,20]
plt.barh(categories,values)
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart')
plt.show()
```

```python
#more than one value stacked bar graph
categories=['Cat A','Cat B','Cat C', 'Cat D']
values1=[25,30,15,20]
values2=[10,12,15,14]
plt.bar(categories,values1,label='value1')
plt.bar(categories,values2,label='value2',bottom=values1 #agr bottom m vlaue 1 dikhani h to 
        )
plt.legend() #shows orange represent value 2 and blue represent value 1
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart')
plt.show()
```

```python
#random number using numpy
import numpy as np
data=np.random.randn(1000) #n is normaly dostributed number
plt.hist(data,bins=25)
plt.xlabel('Value')
plt.ylabel('frequency')
plt.title('histogram')
plt.show()
```

```python
data=np.random.randint(1000) #int givs one number randomly generated
plt.hist(data,bins=25)
plt.xlabel('Value')
plt.ylabel('frequency')
plt.title('histogram')
plt.show()
```

```python
y=[1,4,9,16,25,36,49,64]
x1=[1,16,30,42,55,68,77,88]
x2=[1,6,12,18,28,40,52,65]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1]) #[rightedge,bottomedge,width,height]
l1=ax.plot(x1,y,'ys-')#solid line with yellow color marker
l2=ax.plot(x2,y,'go--')#dashed line with green color marker
ax.legend(labels=('tv','smartphone'),loc='lower right')
ax.set_title("Advewrtisment effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```
