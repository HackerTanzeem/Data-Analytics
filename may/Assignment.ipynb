                                                                   A.

 Converting Sub_metering_1 from object to integer

```python
df['Sub_metering_1'] = pd.to_numeric(df['Sub_metering_1'], errors='coerce')
df['Sub_metering_1'] = df['Sub_metering_1'].fillna(0).astype(int)
```

Checking the value

```python
df.info()
```

Converting Sub_metering_2 from object to integer

```python
df['Sub_metering_2'] = pd.to_numeric(df['Sub_metering_2'], errors='coerce')
df['Sub_metering_2'] = df['Sub_metering_2'].fillna(0).astype(int)
```

```python
df.info()
```

```python
df['Total_Sub_metering'] = df['Sub_metering_1'] + df['Sub_metering_2'] + df['Sub_metering_3'].fillna(0)
```

Making a new column named "Total_Sub_metering" and giving the summation of sub meterings as its data

                                                                              B.

Converting Date object type to datetime type

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
```

```python
df.info()
```

                                                                    C.

Splitinng Date column into Date,Month and Year

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df.info()
```

```python
df
```

Making Time from object type to datetime type

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')
```

```python
df.info()
```

Spliting Time column into Hour,Minute and Second

```python
df['Hour'] = df['Time'].dt.hour
df['Minute'] = df['Time'].dt.minute
df['Second'] = df['Time'].dt.second
```

```python
df
```

                                                                        D.

Checking for null datas

```python
df.isnull()
```

By using .sum() function we can se the columns having total number of null value

```python
df.isnull().sum()
```

Removing null values with the mean value of the respective column

```python
mean_value = df['Sub_metering_3'].mean()
df['Sub_metering_3'] = df['Sub_metering_3'].fillna(mean_value)
```

```python
df.isnull().sum()
```
